```python
from mapboxgl.utils import *
from mapboxgl.viz import *
from tools.helpers import read_json
```

```python
config = read_json('config.json')
mapbox_key = config['mapbox_key']
```

```python
# Just show a map of the data
viz = CircleViz('data/processing/processing.geojson', 
                access_token=mapbox_key, 
                radius=1, 
                center=(-95, 40), 
                zoom=3)
viz.show()
```

```python
# Generate data breaks using numpy quantiles and color stops from colorBrewer
measure = 'num'
color_breaks = [round(dfObj[measure].quantile(q=x*0.1), 2) for x in range(1, 9)]
color_stops = create_color_stops(color_breaks, colors='YlGnBu')

# Create the viz from the dataframe
viz = CircleViz('data/processing.geojson',
                access_token=mapbox_key, 
                color_property="num",
                color_stops=color_stops,
                radius=2.5,
                stroke_color='black',
                stroke_width=0.2,
                center=(-95, 40),
                zoom=3,
                below_layer='waterway-label')
viz.show()
```
